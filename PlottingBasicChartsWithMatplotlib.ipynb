 Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

covid = pd.read_csv('owid-covid-data.csv')
italyDf = covid[covid['location'].isin(['Italy'])]

plt.figure(figsize=(10, 6))

plt.subplot(1,3,1)
plt.hist(italyDf.total_cases,bins = 50, color="#FF914D")
plt.title('Daily Deaths for Italy',fontsize= 15)
plt.ylabel("deaths",fontstyle = 'normal', 
          fontsize = 20)
plt.xlabel("days",fontstyle = 'normal', 
          fontsize = 20)

plt.subplot(1,3,2)
plt.hist(italyDf.total_cases,bins = 10, color="#008080")
plt.title('Daily Deaths for Italy', fontsize =15)
plt.ylabel("deaths",fontstyle = 'normal', 
          fontsize = 20)
plt.xlabel("days",fontstyle = 'normal', 
          fontsize = 20)

plt.subplot(1,3,3)
plt.hist(italyDf.total_cases,bins = 100, color="#800000")
plt.title('Daily Deaths for Italy',fontsize = 15)
plt.ylabel("deaths",fontstyle = 'normal', 
          fontsize = 20)
plt.xlabel("days",fontstyle = 'normal',
          fontsize = 20)

plt.tight_layout()
plt.show()
```

Plot a scatter plot of new cases and new death for Germany and France.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

covid = pd.read_csv('owid-covid-data.csv')
plt.figure(figsize=(10, 6))

plt.title('new cases and new death')
for locate in ['Germany', 'France']:
  plt.scatter(covid[covid.location == locate].new_cases, 
               covid[covid.location == locate].new_deaths, label = locate)

plt.xlabel("new cases")
plt.ylabel("new deaths")
plt.legend()
plt.show()
```

Plot a boxplot for daily deaths for any country you choose.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

covid = pd.read_csv('owid-covid-data.csv')
plt.figure(figsize=(10, 6))


turkey = covid[covid['location'].isin(['Turkey'])]

plt.title('Daily Deaths For Turkey', fontsize=25)
plt.boxplot(turkey['new_deaths'].dropna())
plt.xlabel('days',fontstyle = 'italic', 
          fontsize = 20)
plt.ylabel('deaths',fontstyle = 'italic', 
          fontsize = 20)
plt.show()
#new_deaths
```

Calculate the total case for each continent and plot a pie chart



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

covid = pd.read_csv('owid-covid-data.csv')
plt.figure(figsize=(10, 6))


continents = covid[['continent','total_cases']]
sonuc = continents.groupby(by=['continent'])['total_cases'].sum()
plt.pie(sonuc,labels = sonuc.index,autopct='%1.1f%%',shadow=True,startangle=90)
plt.show()
sonuc
```
